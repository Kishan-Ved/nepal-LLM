```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
# driver = webdriver.Chrome()
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--disable-css")
chrome_options.add_argument("--blink-settings=imageEnabled=false")
driver = webdriver.Chrome(options=chrome_options)
print("Running in headless mode.")
```

```python
links = pd.read_csv("./online-khabar.csv")
```

```python
iit = 9860
```

```python
import os
from selenium.webdriver.common.by import By
from selenium.webdriver.remote.webelement import WebElement
from selenium.common.exceptions import NoSuchWindowException
from IPython.display import display, Javascript

def rerun_cell():
    display(Javascript('IPython.notebook.execute_cell()'))

# Create directory if it doesn't exist
path = "./online-khabar-articles"
if not os.path.exists(path):
    os.mkdir(path)

for ii in range(iit, 14000):
    try:
        print(ii)
        link = links["url"][ii]
        driver.get(link)

        content = driver.find_elements(By.CLASS_NAME, 'ok-single-middle')
        if len(content) < 1:
            print(f"Some problem with link: {link}")
            continue
        
        # Process article header
        art = content[0].find_elements(By.CLASS_NAME, 'ok-post-title-right')
        if len(art) < 1:
            print(f"Some problem with link: {link}")
            continue

        header = art[0].find_elements(By.TAG_NAME, 'h1')
        header_text = ""
        if len(header) > 0:
            header_text = header[0].get_attribute("innerText")
        
        # Process description blocks
        txt = content[0].find_elements(By.CLASS_NAME, "ok18-single-post-content-wrap")
        txt_str = ""
        for i in range(len(txt)):
            p_tag = txt[i].find_elements(By.TAG_NAME, 'p')
            for j in range(len(p_tag)):
                txt_str = txt_str + p_tag[j].get_attribute("innerText")+" "
        
        # Save to file
        file_name = str(ii) + ".txt"
        file_path = os.path.join(path, file_name)
        with open(file_path, 'w') as f:
            f.write(header_text)
            f.write("\n\n")
            f.write("\n\n")
            f.write(txt_str)
            
        ii += 1
        iit = ii

    except NoSuchWindowException as e:
        print("NoSuchWindowException caught, rerunning the cell...")
        iit += 1
        rerun_cell()  # This will automatically rerun the current cell
```

```python
# Close the WebDriver session
driver.quit()
```
