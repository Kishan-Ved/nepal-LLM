```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

# Read links from the CSV file
links = pd.read_csv("./online-khabar.csv")

# Starting point
iit = 20000

# Create directory if it doesn't exist
path = "online-khabar-articles"
if not os.path.exists(path):
    os.mkdir(path)

for ii in range(iit, 21000):
    try:
        print(ii)
        link = links["url"][ii]

        # Send an HTTP request to fetch the page content
        response = requests.get(link)
        if response.status_code != 200:
            print(f"Failed to retrieve {link} with status code {response.status_code}")
            continue

        # Parse the content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Process the article header
        art = soup.find('div', class_='ok-post-title-right')
        header_text = ""
        if art:
            header = art.find('h1')
            if header:
                header_text = header.get_text(strip=True)

        # Process description blocks
        txt = soup.find('div', class_="ok18-single-post-content-wrap")
        txt_str = ""
        if txt:
            paragraphs = txt.find_all('p')
            for p in paragraphs:
                txt_str += p.get_text(strip=True) + " "

        if not header_text and not txt_str:
            print(f"No content found for link: {link}")
            continue

        # Save to file
        file_name = f"{ii}.txt"
        file_path = os.path.join(path, file_name)
        with open(file_path, 'w', encoding='utf-8') as f:
            f.write(header_text)
            f.write("\n\n")
            f.write(txt_str)

        # Update the iterator
        ii += 1
        iit = ii

    except Exception as e:
        print(f"Exception occurred for link {link}: {e}")
        iit += 1
        continue  # Move to the next link if there's an exception

print("Scraping completed.")
```
