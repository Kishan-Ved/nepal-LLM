```python
chrome_options = Options()
prefs = {
    "profile.managed_default_content_settings.images": 2,  # Disable images
    "profile.managed_default_content_settings.stylesheets": 2,  # Disable CSS
    "profile.default_content_setting_values.notifications": 2  # Disable notifications if needed
}

chrome_options.add_experimental_option("prefs", prefs)

# Initialize the WebDriver with the modified options
driver = webdriver.Chrome(options=chrome_options)
```

```python
from time import sleep, time

total_duration = 30 * 60  # 30 minutes in seconds
interval = 1  # 3 seconds

# Record the start time
start_time = time()

try:
    while time() - start_time < total_duration:
        # Scroll down by a fixed amount
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        
        # Wait for the specified interval
        sleep(interval)
        
        # Check if reached the bottom of the page
        current_position = driver.execute_script("return window.pageYOffset + window.innerHeight;")
        page_height = driver.execute_script("return document.body.scrollHeight;")
        
        # If at the bottom, scroll back up and continue
        if current_position >= page_height:
            driver.execute_script("window.scrollTo(0, 0);")
            sleep(interval)
except KeyboardInterrupt:
    print("Scrolling interrupted by user.")
```

#### Buttons

##### Getting specific section/division containing page numbers.

```python
next_button = driver.find_elements(By.CSS_SELECTOR, '[data-testid="pagination-next-button"]')
if len(next_button) != 1:
    print(f"Some problem with link: {base_url}")
```

```python
next_button[0].click()
```

#### Link Extraction

##### Getting specific section/division tag. \<div>

```python
div1 = driver.find_elements(By.CLASS_NAME, "teaser.offset")
if len(div1) < 1:
    print(f"Some problem with link: {base_url}")
```

##### Extracting all anchor tag \<a> in the division

```python
link = []
for i in range(len(div1)):
    tag_a = div1[i].find_elements(By.TAG_NAME, 'a')
    if len(tag_a) < 1:
        print(f"Some problem with link: {base_url}")
    for j in range(len(tag_a)):
        link.append(tag_a[j].get_attribute("href"))
```
